## kayıp veri sorunun halledilmesi
###### Verinin kayıp olduğu kayıtları silmek.
###### Kayıp olan verilerin el ile doldurulması.
###### Tüm kayıp verilere aynı değeri vermek.
###### Kayıp olan verilere merkezi eğilim ölçülerinden birini girmek (ortalama, ortanca veya mod)
###### Diğer değişkenlerden yararlanarak değerin tahminedilmesi.

```python
#kayıp veri olan sütunların silinmesi
veri2=veri.drop(labels=["bds","talasemi"],axis=1)
print(veri2.info())
```

```python
#kayıp olan satırların silinmesi
veri3=veri.dropna(axis=0)
print(veri3.info())
```

```python
veri.info()
```

```python
impK =SimpleImputer(missing_values=np.nan,strategy="most_frequent")#strategy "mean","constant"
impS =SimpleImputer(missing_values=np.nan,strategy="median")
bds=veri["bds"].values.reshape(-1,1)
tal=veri["talasemi"].values.reshape(-1,1)
veri["bds"]=impS.fit_transform(bds)
veri["talasemi"]=impK.fit_transform(tal)
veri.info()
```

```python
# print(veri.describe())
print(veri["sk"].describe())
print("q1",veri["sk"].describe()["25%"])
print("q3",veri["sk"].describe()["75%"])
# aykırı(gürültülü) değer
# iqr=q3-q1
#alt_sınır=q1-1.5*iqr
#üst_sınır=q3+1.5*iqr
```

```python
sk_q1=veri["sk"].describe()["25%"]
sk_q3=veri["sk"].describe()["75%"]
sk_IQR=sk_q3-sk_q1
alt_sinir=sk_q1-1.5*sk_IQR
ust_sinir=sk_q3+1.5*sk_IQR
aykirilar=veri[(veri["sk"]<alt_sinir)|(veri["sk"]>ust_sinir)]
aykirilar
#veri=veri[(veri["sk"]>alt_sinir)&(veri["sk"]<ust_sinir)]
```

```python
veri.drop(labels=aykirilar.index,axis=0,inplace=True)
veri.info()
```
